```python
import MetaTrader5 as mt5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# If needed:
# !pip install yfinance
import yfinance as yf


# =========================
# 1) CONNECT TO MT5
# =========================
if not mt5.initialize():
    raise RuntimeError(f"MT5 initialize failed: {mt5.last_error()}")

print("MT5 connected:", mt5.last_error())


# =========================
# 2) PULL ACCOUNT INFO
# =========================
account = mt5.account_info()
if account is None:
    mt5.shutdown()
    raise RuntimeError(f"account_info failed: {mt5.last_error()}")

account_dict = account._asdict()
account_df = pd.Series(account_dict, name="account")

# =========================
# 3) PULL OPEN POSITIONS
# =========================
positions = mt5.positions_get()
positions_df = pd.DataFrame([p._asdict() for p in positions]) if positions else pd.DataFrame()

if not positions_df.empty:
    if "time" in positions_df.columns:
        positions_df["time"] = pd.to_datetime(positions_df["time"], unit="s")

print("\nOPEN POSITIONS")
if not positions_df.empty:
    display(positions_df[["symbol", "type", "volume", "price_open", "price_current", "profit", "time"]])
else:
    display(positions_df)


# =========================
# 4) PULL RECENT DEALS
# =========================
date_to = datetime.now()
date_from = date_to - timedelta(days=90)

deals = mt5.history_deals_get(date_from, date_to)
deals_df = pd.DataFrame([d._asdict() for d in deals]) if deals else pd.DataFrame()

if not deals_df.empty:
    if "time" in deals_df.columns:
        deals_df["time"] = pd.to_datetime(deals_df["time"], unit="s")

# Keep only instrument-related rows
if not deals_df.empty and "symbol" in deals_df.columns:
    trade_deals_df = deals_df[deals_df["symbol"].notna() & (deals_df["symbol"] != "")].copy()
else:
    trade_deals_df = pd.DataFrame()

print("\nRECENT DEALS")
if not trade_deals_df.empty:
    display(
        trade_deals_df[
            ["time", "symbol", "type", "entry", "volume", "price", "profit", "commission", "swap", "position_id"]
        ].sort_values("time")
    )
else:
    display(trade_deals_df)


# =========================
# 5) BUILD CURRENT POSITION SUMMARY
# =========================
if positions_df.empty:
    mt5.shutdown()
    raise ValueError("No open positions found. This notebook version expects at least one open position.")

summary = pd.DataFrame({
    "symbol": [positions_df["symbol"].iloc[0]],
    "n_positions": [len(positions_df)],
    "total_volume": [positions_df["volume"].sum()],
    "avg_entry_price": [((positions_df["price_open"] * positions_df["volume"]).sum() / positions_df["volume"].sum())],
    "current_price": [positions_df["price_current"].iloc[0]],
    "floating_pnl": [positions_df["profit"].sum()]
})

print("\nCURRENT POSITION SUMMARY")
display(summary)


# =========================
# 6) GET DAILY BARS FROM MT5
# =========================
symbol = summary.loc[0, "symbol"]
entry_time = positions_df["time"].min().normalize()

rates = mt5.copy_rates_from_pos(symbol, mt5.TIMEFRAME_D1, 0, 120)
if rates is None:
    mt5.shutdown()
    raise RuntimeError(f"copy_rates_from_pos failed for {symbol}: {mt5.last_error()}")

rates_df = pd.DataFrame(rates)
if rates_df.empty:
    mt5.shutdown()
    raise ValueError(f"No daily bars returned for {symbol}")

rates_df["time"] = pd.to_datetime(rates_df["time"], unit="s")
rates_df = rates_df.sort_values("time").reset_index(drop=True)

# Keep bars from a few days before entry onward
rates_df = rates_df[rates_df["time"] >= entry_time - pd.Timedelta(days=5)].copy()

print("\nDAILY BARS")
display(rates_df.tail(10))


# =========================
# 7) GET CONTRACT POINT VALUE
# =========================
symbol_info = mt5.symbol_info(symbol)
if symbol_info is None:
    mt5.shutdown()
    raise RuntimeError(f"symbol_info failed for {symbol}: {mt5.last_error()}")

tick_size = symbol_info.trade_tick_size
tick_value = symbol_info.trade_tick_value

if tick_size is None or tick_size == 0:
    mt5.shutdown()
    raise ValueError(f"Invalid tick_size for {symbol}: {tick_size}")

point_value_per_lot = tick_value / tick_size

print("\nCONTRACT INFO")
print("symbol:", symbol)
print("tick_size:", tick_size)
print("tick_value:", tick_value)
print("point_value_per_lot:", point_value_per_lot)


# =========================
# 8) RECONSTRUCT BALANCE + EQUITY
# =========================
balance_now = account.balance
equity_now = account.equity

total_volume = float(summary.loc[0, "total_volume"])
avg_entry_price = float(summary.loc[0, "avg_entry_price"])

# Realized cashflow impact from deals
if not trade_deals_df.empty:
    trade_deals_df["cashflow"] = (
        trade_deals_df["profit"].fillna(0.0)
        + trade_deals_df["commission"].fillna(0.0)
        + trade_deals_df["swap"].fillna(0.0)
    )
    daily_cashflows = (
        trade_deals_df.groupby(trade_deals_df["time"].dt.normalize())["cashflow"]
        .sum()
        .sort_index()
    )
    total_cashflows = trade_deals_df["cashflow"].sum()
else:
    daily_cashflows = pd.Series(dtype=float)
    total_cashflows = 0.0

# Infer initial capital before these strategy cashflows
initial_capital = balance_now - total_cashflows

report_df = rates_df[["time", "close"]].copy()
report_df = report_df.rename(columns={"time": "date", "close": "es_close"})
report_df["date"] = report_df["date"].dt.normalize()

report_df["realized_cashflow"] = report_df["date"].map(daily_cashflows).fillna(0.0)
report_df["balance"] = initial_capital + report_df["realized_cashflow"].cumsum()

# Mark-to-market unrealized PnL from entry date onward
report_df["unrealized_pnl"] = np.where(
    report_df["date"] >= entry_time,
    total_volume * point_value_per_lot * (report_df["es_close"] - avg_entry_price),
    0.0
)

report_df["equity"] = report_df["balance"] + report_df["unrealized_pnl"]

print("\nREPORT DF")
display(report_df.tail(10))


# =========================
# 9) DOWNLOAD SPY FROM YAHOO FINANCE
# =========================
start_date = report_df["date"].min().date()
end_date = (report_df["date"].max() + pd.Timedelta(days=2)).date()

spy = yf.download("SPY", start=start_date, end=end_date, auto_adjust=True, progress=False)

if spy.empty:
    mt5.shutdown()
    raise ValueError("No SPY data downloaded from Yahoo Finance")

# Handle both flat and multi-index column cases
spy = spy.copy()
if isinstance(spy.columns, pd.MultiIndex):
    spy.columns = [c[0] for c in spy.columns]

spy_df = spy.reset_index()

# Normalize the first column name to 'date'
first_col = spy_df.columns[0]
spy_df = spy_df.rename(columns={first_col: "date"})

if "Close" not in spy_df.columns:
    mt5.shutdown()
    raise ValueError(f"SPY data does not contain 'Close'. Columns: {spy_df.columns.tolist()}")

spy_df = spy_df[["date", "Close"]].rename(columns={"Close": "spy_close"})
spy_df["date"] = pd.to_datetime(spy_df["date"]).dt.normalize()

print("\nSPY DATA")
display(spy_df.tail(10))


# =========================
# 10) MERGE AND BUILD INDEX SERIES
# =========================
report_df = report_df.merge(spy_df, on="date", how="left")
report_df["spy_close"] = report_df["spy_close"].ffill()

# Drop rows before SPY exists, if any
report_df = report_df.dropna(subset=["spy_close"]).reset_index(drop=True)

report_df["account_equity_index"] = report_df["equity"] / report_df["equity"].iloc[0]
report_df["account_balance_index"] = report_df["balance"] / report_df["balance"].iloc[0]
report_df["spy_index"] = report_df["spy_close"] / report_df["spy_close"].iloc[0]

print("\nNORMALIZED PERFORMANCE")
display(report_df[["date", "account_equity_index", "account_balance_index", "spy_index"]].tail(10))


# =========================
# 11) PERFORMANCE STATS
# =========================
def max_drawdown(index_series: pd.Series) -> float:
    running_max = index_series.cummax()
    drawdown = index_series / running_max - 1.0
    return float(drawdown.min())

equity_ret = report_df["equity"].pct_change().fillna(0.0)
balance_ret = report_df["balance"].pct_change().fillna(0.0)
spy_ret = report_df["spy_close"].pct_change().fillna(0.0)

n_days = len(report_df)

stats = {
    "start_date": report_df["date"].min(),
    "end_date": report_df["date"].max(),
    "n_days": n_days,
    "current_balance": report_df["balance"].iloc[-1],
    "current_equity": report_df["equity"].iloc[-1],
    "current_unrealized_pnl": report_df["unrealized_pnl"].iloc[-1],
    "account_total_return": report_df["account_equity_index"].iloc[-1] - 1.0,
    "balance_total_return": report_df["account_balance_index"].iloc[-1] - 1.0,
    "spy_total_return": report_df["spy_index"].iloc[-1] - 1.0,
    "account_max_drawdown": max_drawdown(report_df["account_equity_index"]),
    "balance_max_drawdown": max_drawdown(report_df["account_balance_index"]),
    "spy_max_drawdown": max_drawdown(report_df["spy_index"]),
}

if n_days > 1:
    stats["account_ann_return"] = float(report_df["account_equity_index"].iloc[-1] ** (252 / n_days) - 1.0)
    stats["balance_ann_return"] = float(report_df["account_balance_index"].iloc[-1] ** (252 / n_days) - 1.0)
    stats["spy_ann_return"] = float(report_df["spy_index"].iloc[-1] ** (252 / n_days) - 1.0)

    stats["account_ann_vol"] = float(equity_ret.std() * np.sqrt(252))
    stats["balance_ann_vol"] = float(balance_ret.std() * np.sqrt(252))
    stats["spy_ann_vol"] = float(spy_ret.std() * np.sqrt(252))

    stats["account_sharpe"] = float(equity_ret.mean() / equity_ret.std() * np.sqrt(252)) if equity_ret.std() != 0 else np.nan
    stats["balance_sharpe"] = float(balance_ret.mean() / balance_ret.std() * np.sqrt(252)) if balance_ret.std() != 0 else np.nan
    stats["spy_sharpe"] = float(spy_ret.mean() / spy_ret.std() * np.sqrt(252)) if spy_ret.std() != 0 else np.nan

stats_df = pd.Series(stats, name="value").to_frame()

print("\nSTATS")
display(stats_df)


# =========================
# 12) PLOT CUMULATIVE PERFORMANCE
# =========================
plt.figure(figsize=(12, 6))
plt.plot(report_df["date"], report_df["account_equity_index"], label="Account Equity")
plt.plot(report_df["date"], report_df["account_balance_index"], label="Account Balance")
plt.plot(report_df["date"], report_df["spy_index"], label="SPY")
plt.title("Cumulative Performance")
plt.xlabel("Date")
plt.ylabel("Normalized Index")
plt.grid(True)
plt.legend()
plt.show()


# =========================
# 13) PLOT DRAWDOWNS
# =========================
report_df["account_drawdown"] = report_df["account_equity_index"] / report_df["account_equity_index"].cummax() - 1.0
report_df["balance_drawdown"] = report_df["account_balance_index"] / report_df["account_balance_index"].cummax() - 1.0
report_df["spy_drawdown"] = report_df["spy_index"] / report_df["spy_index"].cummax() - 1.0

plt.figure(figsize=(12, 5))
plt.plot(report_df["date"], report_df["account_drawdown"], label="Account Equity DD")
plt.plot(report_df["date"], report_df["balance_drawdown"], label="Account Balance DD")
plt.plot(report_df["date"], report_df["spy_drawdown"], label="SPY DD")
plt.title("Drawdowns")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.grid(True)
plt.legend()
plt.show()


# =========================
# 14) SHOW FINAL TABLES
# =========================
print("\nOPEN POSITIONS VIEW")
if not positions_df.empty:
    open_positions_view = positions_df[
        ["symbol", "type", "volume", "price_open", "price_current", "profit", "time"]
    ].copy()
    display(open_positions_view)
else:
    display(positions_df)

print("\nRECENT DEALS VIEW")
if not trade_deals_df.empty:
    recent_deals_view = trade_deals_df[
        ["time", "symbol", "type", "entry", "volume", "price", "profit", "commission", "swap", "position_id"]
    ].sort_values("time", ascending=False).copy()
    display(recent_deals_view)
else:
    display(trade_deals_df)


# =========================
# 15) OPTIONAL CONSISTENCY CHECK
# =========================
print("\nCONSISTENCY CHECK")
print("Account snapshot balance:", balance_now)
print("Reconstructed latest balance:", report_df['balance'].iloc[-1])
print("Account snapshot equity:", equity_now)
print("Reconstructed latest equity:", report_df['equity'].iloc[-1])

# Leave MT5 connected if you want to keep working.
# Uncomment this when finished:
# mt5.shutdown()
```
